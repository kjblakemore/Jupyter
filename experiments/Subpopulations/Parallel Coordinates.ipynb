# Frequent Item Sets Visualization of Titanic Survived Subpopulation

```python
import sys
import warnings

import pandas as pd
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt

warnings.filterwarnings("ignore")

DATA_DIR_NAME = '../data/'
```

```python
DATA_SET_NAME = "titanic_survived_subpop_small"
pdf = pd.read_csv(DATA_DIR_NAME + DATA_SET_NAME + ".csv", encoding='latin-1')

sns.set(font_scale=1.5)

g = sns.FacetGrid(pdf, hue="Itemset", palette="Set2", size=12)
g.map(plt.scatter, "Item Rank", "Cumulative Lift", s=100, alpha=.4)
g.map(plt.plot, "Item Rank", "Cumulative Lift", linewidth=4, alpha=.4)
g.set(xticklabels=['', 'Sex=female', 'Pclass=1', 'SibSp=1', 'SibSp=0', 'Parch=0', 'Embarked=S'])
g.set_xlabels("Item")
g.fig.suptitle('Top 7 Frequent Item Sets of Titanioc Survived Subpopulation')
g.fig.subplots_adjust(top=.95)
g.add_legend()
```
